```python
import pickle

with open('/content/drive/My Drive/data/ngrams_recall_precision10.pkl', 'rb') as f:
    ngram_recall, ngram_precision = pickle.load(f)

with open('/content/drive/My Drive/data/transformer_recall_precision10.pkl', 'rb') as f:
    transformer_recall, transformer_precision = pickle.load(f)
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Plot the precision-recall curve
plt.figure(figsize=(8, 6))
plt.plot(ngram_recall, ngram_precision, color='b', label = "ngrams")
plt.plot(transformer_recall, transformer_precision, color='r', label = "transformer")
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()

plt.xlim([0, 0.1])


plt.show()
```
